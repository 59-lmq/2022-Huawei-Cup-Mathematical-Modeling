```python
import os
import pandas as pd
import warnings
from chazhi import *
warnings.filterwarnings('ignore')

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
sns.set_palette("husl") #设置所有图的颜色，使用hls色彩空间
```

```python
# 读取 土壤湿度、土壤蒸发量、植被指数、径流量等附件
data3 = pd.read_excel("data/附件3、土壤湿度2022—2012年.xls")
data3.head()
```

```python
data4 = pd.read_excel("data/附件4、土壤蒸发量2012—2022年.xls")
data4.head()
```

```python

data = pd.merge(data3, data4)
data.head()
```

```python
temp=pd.DataFrame()
for i in os.walk('./data/附件8'):
    for k in i[2]:
        temp=pd.concat([temp,pd.read_excel('./data/附件8/%s'%k)])
temp
```

```python
data = pd.merge(data, temp)
data.isnull().sum()/data.shape[0]
```

```python
del data['积雪深度(mm)']
del data['平均最大瞬时风速(knots)']
del data['最大瞬时风速极值(knots)']
del data['经度']
del data['纬度']
del data['站点号']
del data['经度(lon)']
del data['纬度(lat)']
del data['海拔高度(m)']
```

```python
data.isnull().sum()/data.shape[0]
```

```python
for i in data['月份'].unique():
    display(data[data['月份']==i])
```

```python
data = filter_extreme_3sigma(data)
```

```python

# month_idx = 1
for month_idx in range(12):
    month_idx += 1
    month_data = data[data['月份']==month_idx]
    month_data = month_data.sort_values("年份")
    month_data.reset_index(inplace=True, drop=True)
    month_data.to_csv('P2_1_data/month_data_%d.csv'%month_idx)

    for i in ['10cm湿度(kg/m2)', '40cm湿度(kg/m2)',
              '100cm湿度(kg/m2)', '200cm湿度(kg/m2)']:
        plt.plot(month_data['年份'].values, month_data[i].values, label=i)

    plt.grid()
    plt.legend(loc='best')
    plt.xlabel('年份')
    plt.ylabel("湿度")
    plt.title("2012~2022 %d 月份湿度情况"%month_idx)
    plt.savefig("P2_1_figure/2012~2022 %d 月份湿度情况.jpg"%month_idx, dpi=600)
    plt.show()
```
