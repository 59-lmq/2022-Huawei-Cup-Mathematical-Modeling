```python
import os
import pandas as pd
import torch

from P2_utils.data_pro import *
from P2_utils.Mo import *
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_palette("husl") #设置所有图的颜色，使用hls色彩空间

%matplotlib inline
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 基础设置
data_list = ['10cm湿度(kg/m2)', '40cm湿度(kg/m2)',
             '100cm湿度(kg/m2)', '200cm湿度(kg/m2)',
             '土壤蒸发量(W/m2)', '土壤蒸发量(mm)',
             '降水量(mm)']
scaler = MinMaxScaler(feature_range=(0, 1))
# 参数设置
seq_length=1 # 时间步长
# input_size=6
input_size = len(data_list)
num_layers=2
hidden_size=6
batch_size=2
n_iters = 16000
lr=0.001
# output_size=6
output_size = len(data_list)

model_save_path = 'P2_2_model_save/'
train_loss_save_path = 'P2_2_loss_csv_save/'
picture_save_path = 'P2_2_lstm_figure/'
csv_save_path = 'P2_2_data/'
```

```python
def train(model):
    step_t = 0
    model.train()
    for i, (batch_x, batch_y) in enumerate(train_loader):
        outputs = model(batch_x)
        optimizer.zero_grad()
        loss = criterion(outputs, batch_y)
        loss.backward()
        optimizer.step()
        # score(batch_y.detach().numpy(), outputs.detach().numpy())
        rmse, mae, r2 = score(batch_y.detach().numpy(), outputs.detach().numpy())
        # print(rmse, mae, r2)
        step_t += 1
        return rmse, mae, r2, step_t, loss.item()


# 获取数据
for month_idx in range(1, 13):
    print("第 %d 个月正开始" % month_idx)
    data_path = 'P2_1_data/month_data_%d.csv' % month_idx
    # print(data_path)
    origin_data = pd.read_csv(data_path)
    temp_data = origin_data[data_list]
    temp_data.astype('float32')
    dataset_month = scaler.fit_transform(temp_data)
    x_data, y_data = fit_lstm_model(dataset_month, input_size=input_size)
    train_loader, test_loader, num_epochs=data_generator(x_data,y_data,x_data,y_data,n_iters,batch_size)

    # 定义模型
    model = Net(input_size,hidden_size,num_layers,output_size,batch_size,seq_length)
    # 定义MSE损失函数
    criterion = torch.nn.MSELoss()
    # 定义Adam优化器
    optimizer = torch.optim.Adam(model.parameters(),lr=lr)
    print(model)

    # 开始训练
    train_dict = {
        'epoch': [],
        'loss': [],
        'rmse': [],
        'mae': [],
        'r2': [],
    }

    step = 0
    for epochs in range(num_epochs):
        # train(epochs)
        _rmse, mae_re, _r2, _step, _loss = train(model)
        step += _step
        if step % 100 == 0:
            # print("epoch: %d, loss: %1.5f" % (epoch, loss.item()))
            print("epoch: %d, loss: %1.5f, RMSE: %1.5f, MAE: %1.5f, R2: %1.5f"
                  % (int(step/100), _loss, _rmse, mae_re, _r2))
            train_dict['epoch'].append(int(step/100))
            train_dict['loss'].append(_loss)
            train_dict['rmse'].append(_rmse)
            train_dict['mae'].append(mae_re)
            train_dict['r2'].append(_r2)
        if mae_re < 0.01:
            # 保存模型
            model_name = data_path.split('/')[1].split('.')[0] + '.h5'
            model_name = os.path.join(model_save_path, model_name)
            torch.save(model.state_dict(), model_name)
            # print("MAE 小于 0.01 模型保存完毕，位于", model_name)
    model_name = data_path.split('/')[1].split('.')[0] + '_final.h5'
    model_name = os.path.join(model_save_path, model_name)
    torch.save(model.state_dict(), model_name)
    print("第 %d 个月的 LSTM 模型训练完毕，保存完毕，位于 %s" % (month_idx, model_name))
    # 保存过程中的各种指标
    train_pdDataframe = pd.DataFrame(train_dict)
    train_loss_csv_name = os.path.join(train_loss_save_path,  data_path.split('/')[1].split('.')[0] + '_loss.csv')
    train_pdDataframe.to_csv(train_loss_csv_name)
    print("第 %d 个月的 LSTM 模型训练完毕, 各种指标保存完毕，位于 %s" % (month_idx, train_loss_csv_name))

    # 获取图片
    years = origin_data["年份"].values
    years = years[1:]

    y_predict_tensor = model(x_data)
    y_predict = scaler.inverse_transform(y_predict_tensor.detach().numpy())
    y_ground = scaler.inverse_transform(y_data.detach().numpy())


    # 保存数据
    years_dict = {'年份': years}
    years_pdDataframe = pd.DataFrame(years_dict)
    y_predict_pdDataframe = pd.DataFrame(y_predict)
    y_predict_pdDataframe.columns = ['预测值' + d for d in data_list]
    y_predict_save = pd.concat([years_pdDataframe, y_predict_pdDataframe], axis=1)

    csv_save_name = os.path.join(csv_save_path, data_path.split('/')[1].split('.')[0] + '_predict.csv')
    y_predict_save_final = pd.merge(origin_data, y_predict_save)
    y_predict_save_final.to_csv(csv_save_name)
    print("数据保存成功，在", csv_save_name)


    # 显示单月，某个湿度的
    for idx in range(len(data_list)):
        title_text = '2012~2022 %d 月份 %s' % (month_idx, data_list[idx])
        print("这是图", title_text)
        plt.plot(years, y_ground[:, idx], color='blue', label='真实数据')
        plt.plot(years, y_predict[:, idx], color='red', label='LSTM预测数据')

        plt.grid()
        plt.xlabel("年份")
        plt.ylabel("湿度")
        plt.title(title_text)
        plt.legend(loc='best')

        figure_name = os.path.join(picture_save_path, title_text.split('(')[0] + '.jpg')
        plt.savefig(figure_name, dpi=600)

        plt.show()

    # 显示单月所有湿度条件
    for hum_idx in range(4):
        plt.plot(years, y_ground[:, hum_idx], label=data_list[hum_idx].split('cm')[0] + ' 真实数据')
        plt.plot(years, y_predict[:, hum_idx], label=data_list[hum_idx].split('cm')[0] + ' LSTM预测数据')

    plt.grid()
    plt.legend(loc='upper left')
    plt.xlabel('年份')
    plt.ylabel("湿度")

    title_text = "2012~2022 %d 月份湿度预测情况" % month_idx
    print("这是图", title_text)
    plt.title(title_text)

    figure_name = os.path.join(picture_save_path, title_text + '.jpg')
    plt.savefig(figure_name, dpi=600)
    plt.show()
    print("第 %d 个月已结束" % month_idx)
```
